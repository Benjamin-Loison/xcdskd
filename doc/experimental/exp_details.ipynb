### Magnetic Field Distortions

SEMs with immersion lenses have significant magnetic fields in the region near the pole piece, including between sample and phosphor screen. These fields distort the trajectories of the electrons scattered from the sample. Kikuchi patterns measured under these conditions need to be "unwarped" to provide correct crystallographic information in the sample coordinate system. 

```python
display(Image(filename='./effects_gallery/mf.png', width=320))
display(Image(filename='./effects_gallery/mf_unwarped.png', width=320))
```
